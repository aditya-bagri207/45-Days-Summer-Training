```python
import matplotlib.pyplot as plt
```

```python
subjects = ["Math", "Science", "English", "Computer"]

marks = [85, 90, 78, 95]

students = ["Aditya", "Rahul", "Aman", "Riya"]

student_marks = [88, 72, 91, 80]
```

```python
plt.figure(figsize=(10, 8))
```

```python
plt.subplot(2, 2, 1)

plt.plot(subjects, marks, marker="o")

plt.title("Line Plot")
```

```python
plt.subplot(2, 2, 2)

plt.bar(students, student_marks)

plt.title("Bar Chart")
```

```python
plt.subplot(2, 2, 3)

plt.hist(student_marks)

plt.title("Histogram")
```

```python
plt.subplot(2, 2, 4)

plt.pie(
    student_marks,
    labels=students,
    autopct="%1.1f%%"
)

plt.title("Pie Chart")
```

```python
plt.tight_layout()
```
