```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python


data = {
    "Age": np.random.randint(22, 50, 50),
    "Salary": np.random.randint(25000, 100000, 50),
    "Department": np.random.choice(
        ["IT", "HR", "Sales", "Finance"],
        50
    ),
    "Experience": np.random.randint(1, 15, 50)
}

df = pd.DataFrame(data)

print(df.head())
```

```python
sns.scatterplot(
    data=df,
    x="Age",
    y="Salary",
    hue="Department"
)

plt.title("Age vs Salary")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
```

```python
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.barplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Average Salary by Department")
plt.show()
```
