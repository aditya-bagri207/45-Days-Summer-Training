```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Math": [85, 90, 78, 95, 88],
    "Science": [80, 85, 75, 92, 89],
    "English": [70, 88, 82, 90, 85],
    "Computer": [95, 98, 90, 99, 97]
}

df = pd.DataFrame(data)

print(df)
```

```python
df["Math"].plot(kind="hist")

plt.title("Math Marks Distribution")
plt.xlabel("Marks")

plt.show()
```

```python
df.plot(kind="box")

plt.title("Box Plot")

plt.show()
```

```python
category = pd.Series({
    "A": 40,
    "B": 30,
    "C": 20,
    "D": 10
})
```

```python
category.plot(kind="pie", autopct="%1.1f%%")

plt.title("Grade Distribution")

plt.ylabel("")

plt.show()
```
